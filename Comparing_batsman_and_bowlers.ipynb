# z-score value for Batsman

```python
total_runs=df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
balls=df.groupby('batsman')['batsman_runs'].count().reset_index()
```

```python
batsman_df=total_runs.merge(balls,on='batsman')
```

```python
batsman_df.rename(columns={'batsman_runs_x':'total_runs','batsman_runs_y':'balls'},inplace=True)
```

```python
batsman_df.head()
```

```python
batsman_df=batsman_df[batsman_df['balls']>200]
```

```python
batsman_df['strike']=(batsman_df['total_runs']/batsman_df['balls'])*100
```

```python
batsman_df.head()
```

```python
batsman_mean=batsman_df['strike'].mean()
batsman_std=batsman_df['strike'].std()
batsman_df['z_score']=(batsman_df['strike']-batsman_mean)/batsman_std
```

```python
batsman_df.head()
```

```python
batsman_df=batsman_df.sort_values('strike',ascending=False)
batsman_df.head()
```

# z-score table for Bowlers

```python
dismissal_kind=['caught','bowled','lbw','caught and bowled','stumped','hit wicket']
wickets_df=df[df['dismissal_kind'].isin(dismissal_kind)]
```

```python
wickets_df.shape
```

```python
wickets=wickets_df.groupby('bowler')['player_dismissed'].count().reset_index()
runs_given=df.groupby('bowler')['total_runs'].sum().reset_index()
bowler_df=wickets.merge(runs_given,on='bowler')
balls=df.groupby('bowler')['total_runs'].count().reset_index()
bowler_df=bowler_df.merge(balls,on='bowler')
```

```python
bowler_df.rename(columns={'player_dismissed':'wickets','total_runs_x':'runs','total_runs_y':'balls'},inplace=True)
```

```python
bowler_df.head()
```

```python
bowler_df=bowler_df[bowler_df['runs']>200]
bowler_df['economy']=100-(bowler_df['runs']/(bowler_df['balls']/6))
```

```python
bowler_df.head()
```

```python
bowler_mean=bowler_df['economy'].mean()
bowler_std=bowler_df['economy'].std()
bowler_df['z_score']=(bowler_df['economy']-bowler_mean)/bowler_std
```

```python
bowler_df=bowler_df.sort_values('z_score',ascending=False)
bowler_df.head()
```

#  function for comparing batsman and bowlersÂ¶


```python
def compare_batsman_bowler(batsman,bowler):
    batsman_flag=False
    bowler_flag=False
    if batsman in list(batsman_df['batsman'].unique()):
        batsman_flag=True
    if bowler in list(bowler_df['bowler'].unique()):
        bowler_flag=True
    if batsman_flag and bowler_flag:
        batsman_info=batsman_df[batsman_df['batsman']==batsman]
        bowler_info=bowler_df[bowler_df['bowler']==bowler]
        batsman_win=0
        bowler_win=0
        if batsman_info['z_score'].values[0]>bowler_info['z_score'].values[0]:
            batsman_win+=1
        elif batsman_info['z_score'].values[0]<bowler_info['z_score'].values[0]:
            bowler_win+=1
        elif batsman_info['z_score'].values[0]==bowler_info['z_sore'].values[0]:
            print(batsman,' and ',bowler,' both have the same performance')
        if batsman_win>bowler_win:
            print(batsman,' is a better player than ',bowler)
        elif batsman_win<bowler_win:
            print(bowler,' is a better player than ',batsman)
    else:
        print('Either one or both the players have not played more than 200 balls. Sorry, cannot compare')
```

```python
compare_batsman_bowler('V Kohli','Harbhajan Singh')
```
